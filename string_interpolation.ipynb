```python
num =45.3
sng='cellar doors'
print("I would like to have" + str(num) +' ' + sng)
```

```python
# string interpolation
print('I would like to have %g %s' % (num,sng))
```

```python
# f-string
print(f'I would like to have {num} and {sng}')
```

```python
for i in range(9,18):
    print(f'{i:.2f} is a number, and {(i/3)**4} is a bigger number')
```
