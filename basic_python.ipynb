```python
import numpy as np
# Fibonaci

def fib(n):
    v=[0,1]
    for i in range(n):
        v.append(sum(v[-2:]))
    return v
```

```python
fib(5)
```

```python
# zip
numberlist=[1,2,3,4,5]
letterlist=['A','B','C','D','E']
for num, lett in zip(numberlist,letterlist):
    print(num, lett)
```

```python
for idx, (num, lett) in enumerate(zip(numberlist,letterlist)):
    print(idx, num, lett)
```
